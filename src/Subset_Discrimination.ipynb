# SUBSET DISCRIMINATION DISCOVERY

This Notebook can be used to detect subset discrimination, by making use of decision-tree classifiers.

The data processing was partially inspired by this [blog post](http://www.valentinmihov.com/2015/04/17/adult-income-data-set) by Valentin Mihov.

```python
# écrire résultat
PRINT_FILE = True
OUTPUT_FILE = None

# only consider the root and tree leaves
LEAVES_ONLY = True

# decision tree attributes
MAX_DEPTH   = 5
MIN_SAMPLES = 100

DATA_DIR="../data/"
GRAPHS_DIR="../graphs/"
REPORTS_DIR="../reports/"

# select dataset
'''
- adultCensus : adult census dataset
- berkeley : berkeley admission data
- simpson : simulated Simpson's paradow
- counterBlack : counter example again transformation from categories to binary (center on black people)
- counterWhite : counter example again transformation from categories to binary (center on white people)
- berkeleyPerso : berkeley admission data but with department as sensitive feature
'''
selectData='counterBlack'
```

```python
def adultCensus():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    # ADULT CENSUS DATASET
    original_data = pd.read_csv(
        DATA_DIR+"adult.data",
        names=[
            "Age", "Workclass", "fnlwgt", "Education", "Education-Num", "Martial Status",
            "Occupation", "Relationship", "Race", "Gender", "Capital Gain", "Capital Loss",
            "Hours per week", "Country", "Target"],
            sep=r'\s*,\s*',
            engine='python',
            na_values="?")
    SENS = "Gender"
    SENS_VALS = ["Gender_Female", "Gender_Male"]
    OUT  = "Target"
    OUT_VALS = ["Target_<=50K", "Target_>50K"]
    OUTPUT = GRAPHS_DIR+"tree_adult.pdf"

def berkeley():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    #BERKELEY ADMISSION DATA
    original_data = pd.read_csv(
        DATA_DIR+"berkeley.data",
        names=[
            "Gender", "Department", "Admitted"],
            sep=r'\s*,\s*',
            engine='python',
            na_values="?")
    SENS = "Gender"
    SENS_VALS = ["Gender_Female", "Gender_Male"]
    OUT  = "Admitted"
    OUT_VALS = ["Admitted_No", "Admitted_Yes"]
    OUTPUT = GRAPHS_DIR+"tree_berkeley.pdf"
    
def simpson():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    # SIMULATED SIMPSON'S PARADOX
    original_data = pd.read_csv(
        DATA_DIR+"credit.data",
        names=[
            "Gender", "Purpose", "Credit"],
            sep=r'\s*,\s*',
            engine='python',
            na_values="?")
    SENS = "Gender"
    SENS_VALS = ["Gender_Female", "Gender_Male"]
    OUT  = "Credit"
    OUT_VALS = ["Credit_No", "Credit_Yes"]
    OUTPUT = GRAPHS_DIR+"tree_simpson.pdf"

    
# Mon test sur le contre-exemple sur la transformation catégorie->binaire (avec X et Y bernoulli 50%)
def counterBlack():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    # compare Noirs et Non Noirs
    original_data = pd.read_csv(
        DATA_DIR+"contre_exemple.data",
        names=["Race", "X", "Y", "Hired"],
        sep=r'\s*,\s*',
        engine='python',
        na_values="?"
        )
    SENS = "Race"
    SENS_VALS = ["Race_White", "Race_Black", "Race_Hispanic"]
    OUT = "Hired"
    OUT_VALS = ["Hired_No", "Hired_Yes"]
    OUTPUT = GRAPHS_DIR+"contre_exemple_black.pdf"
    OUTPUT_FILE = REPORTS_DIR+"contre_exemple_black.txt"
    
def counterWhite():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    #compare Blancs et Non Blancs
    original_data = pd.read_csv(
        DATA_DIR+"contre_exemple.data",
        names=["Race", "X", "Y", "Hired"],
        sep=r'\s*,\s*',
        engine='python',
        na_values="?"
        )
    SENS = "Race"
    SENS_VALS = ["Race_Black", "Race_White", "Race_Hispanic"]
    OUT = "Hired"
    OUT_VALS = ["Hired_No", "Hired_Yes"]
    OUTPUT = GRAPHS_DIR+"contre_exemple_white.pdf"
    OUTPUT_FILE = REPORTS_DIR+"contre_exemple_white.txt"
    
def berkeleyPerso():
    global original_data, SENS, SENS_VALS, OUT, OUT_VALS, OUTPUT, OUTPUT_FILE
    # Mon test sur Berkeley
    original_data = pd.read_csv(
        DATA_DIR+"berkeley.data",
        names=[
            "Gender", "Department", "Admitted"],
            sep=r'\s*,\s*',
            engine='python',
            na_values="?")
    SENS = "Department"
    SENS_VALS = ["Department_A", "Department_B", "Department_C", "Department_D", "Department_E", "Department_F"]
    OUT  = "Admitted"
    OUT_VALS = ["Admitted_No", "Admitted_Yes"]
    OUTPUT = GRAPHS_DIR+"tree_berkeley_perso.pdf"
    
    

{
    'adultCensus': adultCensus,
    'berkeley': berkeley,
    'simpson': simpson,
    'counterBlack': counterBlack,
    'counterWhite': counterWhite,
    'berkeleyPerso': berkeleyPerso,
}[selectData]()

original_data.tail()
```

## Compute the original bias

```python
target = OUT_VALS[1].split('_')[1]
for sens in SENS_VALS:
    sens = sens.split('_')[1]
    print '{}:'.format(sens)
    tot = len(original_data[original_data[SENS]==sens])
    tot_target = len(original_data[(original_data[SENS]==sens) & (original_data[OUT]==target)])
    print '{}, {:.2f}%'.format(tot, (100.0*tot_target)/tot)
```

## Compute bias for married vs unmarried users

For Adult dataset only

```python
if 'Martial Status' in original_data.columns:
    male = original_data[original_data['Gender'] == 'Female']

    tot_married = len(male[male['Martial Status'] == 'Married-civ-spouse'])
    tot_unmarried = len(male[male['Martial Status'] != 'Married-civ-spouse'])

    married_rich = len(male[(male['Martial Status'] == 'Married-civ-spouse') & (male['Target'] == '>50K')])
    unmarried_rich = len(male[(male['Martial Status'] != 'Married-civ-spouse') & (male['Target'] == '>50K')])

    print 'Married:'
    print '{}, {:.2f}%'.format(tot_married, (100.0*married_rich)/tot_married)
    print 'Unmarried:'
    print '{}, {:.2f}%'.format(tot_unmarried, (100.0*unmarried_rich)/tot_unmarried)
```

# Data Preprocessing

## One-hot-encoding

```python
binary_data = pd.get_dummies(original_data)

# Use a single target variable
binary_data[OUT] = binary_data[OUT_VALS[1]]
del binary_data[OUT_VALS[0]]
del binary_data[OUT_VALS[1]]

# DEBUG Bidouille
#binary_data
```

## Train-Test set split, deletion of sensitive attributes, normalization

```python
random.seed(1)

# train-test set split
X_train, X_test, y_train, y_test = cross_validation.train_test_split(binary_data[binary_data.columns.difference([OUT])], binary_data[OUT], train_size=0.25)

# create a test dataset to analyze clusters
data = X_test.copy()
data[OUT] = y_test
data[SENS] = data[SENS_VALS[1]]
'''
# version d'origine
del data[SENS_VALS[0]]
del data[SENS_VALS[1]]
'''

# remove sensitive attributes for the decision-tree learning phase
for sens in SENS_VALS:
    del data[sens]
    del X_train[sens]
    del X_test[sens]
    
'''
# version d'origine
del X_train[SENS_VALS[0]]
del X_train[SENS_VALS[1]]
del X_test[SENS_VALS[0]]
del X_test[SENS_VALS[1]]
'''
    
# build a normalized version of the data for logistic regression
scaler = preprocessing.StandardScaler()
scaler.fit(X_train)
X_train_scaled = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test_scaled = scaler.transform(X_test)
    
print '{} Training Samples, {} Test Samples'.format(len(y_train), len(y_test))

# DEBUG Bidouille
#data
#X_train
```

# Training of Decision-Tree Classifier

```python
cls = tree.DecisionTreeClassifier(criterion='entropy', max_depth=MAX_DEPTH, min_samples_leaf=MIN_SAMPLES, random_state=0)

# baseline logistic regression
cls_base = linear_model.LogisticRegression(random_state=0)

cls.fit(X_train, y_train)
cls_base.fit(X_train_scaled, y_train)
y_pred = cls.predict(X_test)
y_pred_base = cls_base.predict(X_test_scaled)

# compare accuracy of decision tree and logistic regression
print "Tree Classifier:"
print "Accuracy: {:.2f}%".format(100*metrics.accuracy_score(y_test, y_pred))
print "Baseline:"
print "Accuracy: {:.2f}%".format(100*metrics.accuracy_score(y_test, y_pred_base))

# print the tree to a file
dot_data = StringIO()
tree.export_graphviz(cls, out_file=dot_data,feature_names=X_train.columns) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph.write_pdf(OUTPUT) 

# DEBUG Bidouille
#cls.tree_.feature
X_train.columns
```

## Feature Selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import LinearSVC

#binarize features as (below-mean=0, above-mean=1)
binarizer = preprocessing.Binarizer().fit(X_train_scaled)
X_train_bin = pd.DataFrame(binarizer.transform(X_train_scaled), columns=X_train_scaled.columns)
X_test_bin = pd.DataFrame(binarizer.transform(X_test_scaled), columns=X_train_scaled.columns)

# train a l1-penalized logistic regression
cls_l1 = linear_model.LogisticRegression(C=0.01, penalty="l1", dual=False)
cls_l1.fit(X_train_bin, y_train)
y_pred_l1 = map(round,cls_base.predict(X_test_bin))

# compute accuracy of logistic regression
print "Accuracy: {:.2f}".format(100*metrics.accuracy_score(y_test, y_pred_l1))

plt.figure(figsize=(18,8))

# flatten list of regression coeffecients 
#(for some reason it is 1-dimensional for Lasso and 2-dimensional for Logistic-Regression)
cls_l1.coef_ = ndarray.flatten(cls_l1.coef_)

coefs = pd.Series(cls_l1.coef_, index=X_train.columns)
coefs.sort()

coefs.plot(kind="bar")
plt.tight_layout()
plt.show()
```

Select the top features and build clusters

```python
NUM_FEATURES = 5

# sort absolute values of coefficients and select top NUM_FEATURES
abs_coef = abs(coefs)
abs_coef.sort()
features = abs_coef.tail(NUM_FEATURES).index

print 'the {} most relevant features:'.format(NUM_FEATURES)
print coefs[features]

# create new training/test sets with only the selected binary features
X_train_fs = X_train_bin[features]
X_test_fs = X_test_bin[features]

# re-add the mean to obatin a meaningful decision-tree output
for feature in features:
    mean = round(X_train[feature].mean())
    X_train_fs.loc[:,feature] = X_train_fs[feature] + mean
    X_test_fs.loc[:,feature] = X_test_fs[feature] + mean

# decision tree attributes
MIN_SAMPLES = 100
OUTPUT      = GRAPHS_DIR+"tree_fs.pdf"

# learn a decision tree using the selected binary features only
cls = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=MIN_SAMPLES, random_state=0)
cls.fit(X_train_fs, y_train)
y_pred_fs = cls.predict(X_test_fs)

print "Accuracy: {:.2f}".format(100*metrics.accuracy_score(y_test, y_pred_fs))

# print the tree to a file
dot_data = StringIO()
tree.export_graphviz(cls, out_file=dot_data,feature_names=X_train_fs.columns) 
graph = pydot.graph_from_dot_data(dot_data.getvalue()) 
graph.write_pdf(OUTPUT)
```
